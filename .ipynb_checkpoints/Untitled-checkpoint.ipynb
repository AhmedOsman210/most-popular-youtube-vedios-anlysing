```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Most popular 1000 Youtube videos.csv")
df.head(15)
```

```python
print("Missing values:\n", df.isnull().sum())
```

```python
df = df.drop_duplicates()
```

```python
if "Video Views" in df.columns:
    df["Video Views"] = pd.to_numeric(df["Video Views"], errors='coerce')
if "Likes" in df.columns:
    df["Likes"] = pd.to_numeric(df["Likes"], errors='coerce')
if "Dislikes" in df.columns:
    df["Dislikes"] = pd.to_numeric(df["Dislikes"], errors='coerce')
if "Published" in df.columns:
    df["Published"] = pd.to_numeric(df["Published"], errors='coerce')
else:
    print("Column 'Published' not found in the dataset.")
```

```python
print("\nSummary Statistics:\n", df.describe())
```

```python
# Fetch the top 15 most viewed videos
if 'video_views' in df.columns:
    top_15 = df.nlargest(15, 'video_views')  # Get the top 15 most viewed videos
    print("\nTop 15 Most Viewed Videos (Ranked):")
    
    # Display rank, video title, and views in a readable format
    for idx, row in top_15.iterrows():
        print(f"Rank: {idx+1} | Video: {row['video']} | Views: {row['video_views']:,}")
else:
    print("\n'video_views' column not found, please check the dataset.")
```

```python
# Clean the 'likes' column and convert it to numeric values
df['likes'] = df['likes'].replace({',': ''}, regex=True)  # Remove commas if present
df['likes'] = pd.to_numeric(df['likes'], errors='coerce')  # Convert to numeric (invalid values become NaN)

# Fetch the top 15 most liked videos
if 'likes' in df.columns:
    top_liked = df.nlargest(15, 'likes')  # Get the top 15 most liked videos
    print("\nTop 15 Most Liked Videos (Ranked):")
    
    # Display rank, video title, and likes in a readable format
    for idx, row in top_liked.iterrows():
        print(f"Rank: {idx+1} | Video: {row['video']} | Likes: {row['likes']:,}")
else:
    print("\n'likes' column not found, please check the dataset.")
```

```python
# Clean the 'dislikes' column and convert it to numeric values
df['dislikes'] = df['dislikes'].replace({',': ''}, regex=True)  # Remove commas if present
df['dislikes'] = pd.to_numeric(df['dislikes'], errors='coerce')  # Convert to numeric (invalid values become NaN)

# Fetch the top 15 most disliked videos
if 'dislikes' in df.columns:
    top_disliked = df.nlargest(15, 'dislikes')  # Get the top 15 most disliked videos
    print("\nTop 15 Most Disliked Videos (Ranked):")
    
    # Display rank, video title, and dislikes in a readable format
    for idx, row in top_disliked.iterrows():
        print(f"Rank: {idx+1} | Video: {row['video']} | Dislikes: {row['dislikes']:,}")
else:
    print("\n'dislikes' column not found, please check the dataset.")
```

```python
# Group by category and count the number of videos in each category
category_trends = df['category'].value_counts()

# Display the top 10 most common categories
print("Category Trends (Top Categories in the Top 1000 Videos):")
print(category_trends.head(10))
```

```python
# Check the first few entries in the 'published' column
print(df['published'].head())

# Check if there are any invalid entries
invalid_dates = df[df['published'].isna()]
print("Invalid 'published' entries:\n", invalid_dates)

# Clean up the 'published' column (remove invalid rows or correct format)
df['published'] = pd.to_datetime(df['published'], errors='coerce')

# Extract the year from the 'published' column
df['year'] = df['published'].dt.year

# Group by year and count the number of videos per year
yearly_trends = df.groupby('year').size()

# Display the results
print("Yearly Trends: Number of Top Videos per Year")
print(yearly_trends)
```

```python
# Assuming df is your DataFrame
# 1. Bar chart for top categories
plt.figure(figsize=(10,6))
category_counts = df['category'].value_counts()  # Replace 'category' with your actual column name for categories
sns.barplot(x=category_counts.index, y=category_counts.values)
plt.title('Top Categories by Video Count')
plt.xlabel('Category')
plt.ylabel('Number of Videos')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group by year and sum the views
views_by_year = df.groupby('year')['Video views'].sum()

# Find the year with the most viral videos (highest total views)
most_viral_year = views_by_year.idxmax()
most_viral_year_views = views_by_year.max()

print(f"The year with the most viral videos is {most_viral_year} with {most_viral_year_views} total views.")
```

```python
# Calculate the correlation between likes and views
correlation = df['Video views'].corr(df['Likes'])

print(f"The correlation between likes and views is: {correlation}")
```
